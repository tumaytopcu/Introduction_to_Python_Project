#### 7) (10 points) Draw the scatter plot of the dataset using attributes *Weight* and *Height* according to class label *Gender*. Use different colors for *Gender* on the plot.

```python
import matplotlib.pyplot as plt
```

```python
dsm = dataset[dataset.Gender=="Male"]
dsf = dataset[dataset.Gender=="Female"]
x1 = dsm.Weight
x2 = dsf.Weight
y1 = dsm.Height
y2 = dsf.Height
plt.scatter(x1,y1, marker="o")
plt.scatter(x2,y2, marker="s")
plt.xlabel('Weight')
plt.ylabel('Height')
plt.title('Weight and Height by gender')
plt.ylim(150)
plt.show()
```

#### 8) (10 points) Draw the boxplot of the dataset using attributes "Age", "Weight" and "Height" for Female and Male students separately.

```python
z1 = dsm.Age
z2 = dsf.Age

dataset_m = [x1, y1, z1]
dataset_f = [x2, y2, z2]
 
fig = plt.figure()

ax = fig.add_subplot(121)
ax.boxplot(dataset_m) 
ax.set_title('Boxplot of Male Students')
ax.set_xticklabels(['Weight', 'Height', 'Age'])
plt.ylim(0)

ax2 = fig.add_subplot(122)
ax2.boxplot(dataset_f)
ax2.set_title('Boxplot of Female Students')
ax2.set_xticklabels(['Weight', 'Height', 'Age'])
plt.ylim(0)

plt.show()
```
